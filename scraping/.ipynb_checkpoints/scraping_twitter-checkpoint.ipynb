```python
start_time = time.time()

# when any of comments, retweets,likes,views is ZERO, 
# it is blank on the tweet website, so it generates an error, set it to 0 instead
# Since we are using selenium to scrape the data, we can get the text/numbers without any html code,
# I already performed some simple data cleannig in the scraping process as well.
def extract_data(element, xpath):
    try:
        data = element.find_element(By.XPATH, xpath).text
        if not data:
            return 0
        return data
    except:
        return 0
    
    
def get_tweet(element):
    """scrape different features by using XPATH"""
    tweet_id = "None"
    try:
        permalink = element.find_element(By.XPATH, ".//a[contains(@href, '/status/')]").get_attribute('href')
        tweet_id = permalink.split('/')[-1]
        user = element.find_element(By.XPATH, ".//span[contains(text(), '@')]").text
        text = element.find_element(By.XPATH, ".//div[@lang]").text
        comments = extract_data(element, ".//div[@data-testid='reply']//span[@class='css-901oao css-16my406 r-poiln3 r-bcqeeo r-qvutc0']")
        retweets = extract_data(element, ".//div[@data-testid='retweet']//span[@class='css-901oao css-16my406 r-poiln3 r-bcqeeo r-qvutc0']")
        likes = extract_data(element, ".//div[@data-testid='like']//span[@class='css-901oao css-16my406 r-poiln3 r-bcqeeo r-qvutc0']")
        views = extract_data(element, ".//a[contains(@aria-label, 'Views.')]//span[contains(@class, 'css-901oao css-16my406 r-poiln3 r-bcqeeo r-qvutc0')]")
        timestamp = element.find_element(By.XPATH, ".//time").get_attribute("datetime")
        tweets_data = [user, text, comments, retweets, likes, views,timestamp ]
#         print(tweets_data)
    except:
#       return default value when tweet_id||user||text|| timestamp has error.
        tweets_data = ['user', 'text', 0,0,0,0,0]
    return tweet_id, tweets_data

user_data = []
text_data = []
num_comments_list = []
num_retweets_list = []
num_comments_list = []
num_likes_list = []
num_views_list = []
timestamp_list = []
tweet_ids_list = []
tweet_ids_set = set()
scrolling = True
num_of_rows = 300
while scrolling:
    tweets = WebDriverWait(driver, 5).until(
        EC.presence_of_all_elements_located((By.XPATH, "//article[@role='article']")))
    
     # NOTE: ONLY THOSE LOADED IN THE last page will be considered while those from previous page will be forgotten (example: scroll all the way down and then try to find an @username that it's on top --> it won't find it)
    for tweet in tweets[-15:]:
        tweet_id, tweet_list = get_tweet(tweet)
        if tweet_id not in tweet_ids_set:
#           unique tweet id, is used to drop duplicate data
            tweet_ids_set.add(tweet_id)
            tweet_ids_list.append(tweet_id)
            user_data.append(tweet_list[0])
            text_data.append(" ".join(tweet_list[1].split()))
            num_comments_list.append(tweet_list[2])
            num_retweets_list.append(tweet_list[3])
            num_likes_list.append(tweet_list[4])
            num_views_list.append(tweet_list[5])
            timestamp_list.append(tweet_list[6])
            
    # Get the initial scroll height       
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        # Wait to load page
        time.sleep(3)
         # Calculate new scroll height and compare it with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:  # if the new and last height are equal, it means that there isn't any new page to load, so we stop scrolling
            print('height reached')
            scrolling = False
            break
#       break condition based on the number of rows we get works.
        if len(user_data) > num_of_rows: 
            scrolling = False
            break
        else:
            last_height = new_height
            break

print(scrolling)
df_tweets = pd.DataFrame({'tweet_id':tweet_ids_list,
                          'user': user_data, 
                          'text': text_data, 
                          'num_comments': num_comments_list,
                          'num_retweets': num_retweets_list,
                          'num_comments': num_comments_list,
                          'num_views': num_views_list,
                          'timestamp':timestamp_list
                         })
df_tweets.to_csv('technology-@demishassabis-287.5K-0.csv', index=False)
end_time = time.time()
time_taken = end_time - start_time
print(f"Time taken: {time_taken:.2f} seconds")
print(df_tweets)
```

Each tweet data includes name(@xxx),timestamp, tweet content, number of comments, retweets, likes, and views will be scraped.


Tweets with no timestamp will not be scraped
